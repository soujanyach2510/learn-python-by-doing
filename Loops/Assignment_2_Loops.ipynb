```python
# Factorial of any number n is represented by n! and is equal to 1*2*3*....*(n-1)*n. E.g.-
#4! = 1*2*3*4 = 24
#3! = 3*2*1 = 6
#2! = 2*1 = 2
#Also,
#1! = 1
#0! = 1
#Write a program to calculate factorial of a number.

#ans
factorial=1
number=int(input("Enter any number:"))
for i in range(1,number+1):
  factorial=factorial*i
print(factorial)


```

```python
from itertools import count
# Take integer inputs from user until he/she presses q ( Ask to press q to quit after every integer input ).
# Print average and product of all numbers.

#ans

count=0
sum=0
product=1


while(True):
  user_input = input("Enter any number (or 'q' to quit):")
  if user_input.lower() == 'q':
    break
  else:
    number = int(user_input)
    count=count+1
    sum=sum+number
    product=product*number

if count > 0:
  avg=sum/count
  print("Average:",avg)
  print("Product:",product)
else:
  print("No numbers were entered.")
```

```python
# Write a program to find greatest common divisor (GCD) or highest common factor (HCF) of given two numbers.

#ans
a=int(input("Enter first number:"))
b=int(input("Enter second number:"))
gcd = 1
for i in range(1, min(a, b) + 1):
    if a % i == 0 and b % i == 0:
        gcd = i

print("GCD of", a, "and", b, "is", gcd)
```

```python
# A three digit number is called Armstrong number if sum of cube of its digit is equal to number itself.
#E.g.- 153 is an Armstrong number because (13)+(53)+(33) = 153.
# Write all Armstrong numbers between 100 to 500.

#ans

for num in range(100,501):
  num_str=str(num)
  sum_of_cubes=0
  for i in num_str:
    sum_of_cubes=sum_of_cubes+(int(i)**3)

  if sum_of_cubes==num:
    print(num)
```

```python
# Calculate the sum of digits of a number given by user. E.g.-
#INUPT : 123        OUPUT : 6
#INUPT : 12345        OUPUT : 15

# ans:
total_sum_digits=0
number=int(input("Enter any number:"))
while(number>0):
  digit=number%10
  total_sum_digits=total_sum_digits+digit
  number=number/10
  number=int(number)
print(total_sum_digits)
```

```python
# Write a program to find prime factor of a number.
# If a factor of a number is prime number then it is its prime factor.
number=int(input("Enter a number:"))
i = 2
factors = []
while i * i <= number:
    if number % i:
        i += 1
    else:
        number //= i
        factors.append(i)
if number > 1:
    factors.append(number)

print("Prime factors:", factors)
```
