```python
# write a program to check weather a number entered is three digit number or not

# ans
number=input("enter a number")
if len(number)==3:
    print("three digit number")
else:
    print("not a three digit number")
```

```python
# write a program to check weather a person is senior citizen or not?

#ans
age=int(input("enter your age"))
if age>=60:
    print("senior citizen")
else:
    print("not a senior citizen")
```

```python
# write a program to check a weather a number(accepted from user) is divisible by 2 and 3 both

# ans
num=int(input("enter a number"))
if num%2==0 and num%3==0:
    print("The number is divisible by 2 and 3 both")
else:
    print("The number is not divisible by 2 and 3 both")
```

```python
# write a program to find the largest number out of three numbers expected from user

# ans
num1=int(input("enter a number"))
num2=int(input("enter a number"))
num3=int(input("enter a number"))
if num1>num2 and num1>num3:
    print("The largest number is:",num1 )
elif num2>num1 and num2>num3:
    print("The largest number is:",num2)
else:
    print("The largest number is:", num3)
```
