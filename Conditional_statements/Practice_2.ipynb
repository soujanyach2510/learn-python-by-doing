```python
# write a program to accept percentage from user and display the grade according to below condition:

# Mark                      Grade

#>85                          A
#>70 and<=85                  B
#>=60 and <=60                C
# Below 60                    D

# ans
percentage=int(input("enter your percentage"))
if percentage>85:
    print("A")
elif percentage>70 and percentage<=85:
    print("B")
elif percentage>=60 and percentage<=60:
    print("C")
else:
    print("D")
```

```python
# write a program to accept the cost price of a car and display the road tax to be paid according to below condition:

# Cost Price (in Rs)                             Tax
#>1000000                                         10%
# >500000 to <=000000                             5%
# 500000                                          1%

#ans
cost_price=int(input("enter the cost price of a car"))
if cost_price>1000000:
    print("the tax to be paid is:10%")
elif cost_price>500000 and cost_price<=1000000:
    print("the tax to be paid is: 5%")
else:
    print("the tax to be paid is:1%")
```

```python
# write a program to check weather an year is leap year or not

#ans
year=int(input("enter the year"))
if year%4==0:
    print("the year is a leap year")
else:
    print("the year is not a leap year")
```

```python
# write a program to accept a number from 1 to 7 and display the name of day like
# 1 for sunday
# 2 for monday and so on

# ans
number=int(input("enter a number from 1 to 7"))
if number==1:
    print("sunday")
elif number==2:
    print("monday")
elif number==3:
    print("tuesday")
elif number==4:
    print("wednesday")
elif number==5:
    print("thursday")
elif number==6:
    print("friday")
elif number==7:
    print("saturday")
else:
    print("invalid input")
```
