```python
#10.	Write a function min_max_sum(nums) that returns the minimum, maximum, and sum of a list.
def min_max_sum(nums):
  min=nums[0]
  max=nums[0]
  total=0
  for number in nums:
    if number<min:
      min=number
    if number>max:
      max=number
    total=total+number
  return min,max,total
nums=[4,5,1,6,3]
result=(min_max_sum(nums))
print(result)
```

```python
#25.	Write a function apply_if(func, predicate, iterable) that applies a function only to items that pass a condition.
def apply_if(func,predicate,iterable):
  for i in iterable:
    if predicate(i):
      func(i)

def is_odd(x):
  return x%2!=0
def square(x):
  print(x*x)
apply_if(square,is_odd,[1,2,3,4,5])
```

```python
#26.	Write a function curry(func) that converts a two-argument function into a curried version → can be called as f(a)(b).
def curry(func):
  def first_arg(a):
    def second_arg(b):
      return func(a,b)
    return second_arg
  return first_arg
def add(a,b):
  return a+b
curried_add=curry(add)
result=curried_add(5)(10)
print(result)
```

```python
#27.	Write a function variadic_sum(*args, **kwargs) that sums all positional arguments and multiplies by a scale factor if provided.
#kwargs - keword based args like a dictionary, .get method is used to retrieve the values of keys
def variadic_sum(*args, **kwargs):
  total_sum = 0
  for item in args:
    total_sum += item
  scale_factor=kwargs.get("scale",1)
  return total_sum*scale_factor

result1=variadic_sum(1,2,3,scale=2)
print(result1)
result2=variadic_sum(1,2,3)
print(result2)
```

```python
#48.	Create a list of prime pairs (p, q) where both are primes under 100 and p + q is also prime.
def is_prime(num):
  if num<1:
    return False
  for i in range(2,num):
      if num%i==0:
        return False
  return True
prime_numbers=[num for num in range(1,101) if is_prime(num)]
[(p,q) for p in prime_numbers for q in prime_numbers if is_prime(p+q)]
```
