```python
# Write a program to add and multiply two 3x3 matrices.

#ans
A = [
    [1, 2, 3],
    [4, 5, 6],
    [7, 8, 9]
]

B = [
    [9, 8, 7],
    [6, 5, 4],
    [3, 2, 1]
]
add_result= [
    [0, 0, 0],
    [0, 0, 0],
    [0, 0, 0]
]
mul_result= [
    [0, 0, 0],
    [0, 0, 0],
    [0, 0, 0]
]
#Matrix Addition
for i in range(3):
    for j in range(3):
        add_result[i][j] = A[i][j] + B[i][j]

# Matrix Multiplication
for i in range(3):
    for j in range(3):
        for k in range(3):
            mul_result[i][j] += A[i][k] * B[k][j]

# Display Results
print("Matrix A:")
for row in A:
    print(row)

print("\nMatrix B:")
for row in B:
    print(row)

print("\nAddition of A and B:")
for row in add_result:
    print(row)

print("\nMultiplication of A and B:")
for row in mul_result:
    print(row)
```

```python
# student marks
student_marks = [['Name',  ['A','B','C','D','E']],
                 ['Ankit',  [41, 34, 45, 55, 63]],
                 ['Aravind',[42, 23, 34, 44, 53]],
                 ['Lakshay',[32, 23, 13, 54, 67]],
                 ['Gyan',   [23, 82, 23, 63, 34]],
                 ['Pranav', [21, 23, 25, 56, 56]]
                ]
#ðŸ™‡ Who scored the highest marks in the subject B?
student_marks_B=[]
for student in student_marks[1:]:
  name=student[0]
  marks=student[1][1]
  student_marks_B.append([name,marks])
student_marks_B
print(sorted(student_marks_B,key=lambda x:x[1])[-1][-1])

#ðŸ™‡ What is the average marks scored in the subject C?

student_marks_c=[]
for student in student_marks[1:]:
  marks=student[1][2]
  student_marks_c.append(marks)
student_marks_c
average=sum(student_marks_c)/len(student_marks_c)
print(average)

#ðŸ™‡ Who scored the highest percentage of marks?
percentage_students=[]
for student_data in student_marks[1:]:
  name = student_data[0]
  marks_list = student_data[1]
  total_marks = sum(marks_list)
  percentage=(total_marks/500)*100
  percentage_students.append([name,percentage])
print(sorted(percentage_students,key=lambda x:x[1])[-1][-2])

#ðŸ™‡ If considered only top-4 subjects of a candidate, then who scored the highest percentage of marks?

percentage_students_top4=[]
for student_data in student_marks[1:]:
  name = student_data[0]
  marks_list = sorted(student_data[1])[1:]
  total_marks = sum(marks_list)
  percentage=(total_marks/500)*100
  percentage_students_top4.append([name,percentage])
print(sorted(percentage_students_top4,key=lambda x:x[1])[-1][-2])
```
