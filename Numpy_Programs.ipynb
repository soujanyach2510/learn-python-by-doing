```python
#  25.Which NumPy vector operations are legal or illegal?
#Legal Operations
#Element-wise operations on arrays of the same shape
a = np.array([1, 2, 3])
b = np.array([4, 5, 6])
print(a + b)
print(a * b)

#Operations with scalars (broadcasting)

a = np.array([1, 2, 3])
print(a + 10)

#Broadcasting with compatible shapes

a = np.array([[1, 2, 3],
              [4, 5, 6]])   # shape (2,3)
b = np.array([10, 20, 30])  # shape (3,)
print(a + b)



#Comparison & logical operations

a = np.array([1, 2, 3])
b = np.array([2, 2, 2])
print(a > b)
print(np.logical_and(a>0, b<3))

#Matrix multiplication (dot or @)

a = np.array([[1, 2], [3, 4]])
b = np.array([[5, 6], [7, 8]])
print(a @ b)


#2. Illegal Operations

#Element-wise operations on incompatible shapes

a = np.array([1, 2, 3])  # shape (3,)
b = np.array([10, 20])   # shape (2,)
print(a + b)

#Matrix multiplication with incompatible shapes

a = np.array([[1, 2, 3]])   # shape (1,3)
b = np.array([[4, 5], [6, 7]])  # shape (2,2)
print(a @ b)
```

```python
#  44.How do you compute pairwise distances between coordinates?

coords1 = np.array([[0, 0], [1, 1], [2, 0]])
coords2 = np.array([[0, 1], [1, 0]])

# To compute pairwise Euclidean distances, we can use broadcasting
# and the sum of squares.
# (x1 - x2)^2 + (y1 - y2)^2

# Expand dimensions for broadcasting
coords1_expanded = coords1[:, np.newaxis, :]  # shape (3, 1, 2)
coords2_expanded = coords2[np.newaxis, :, :]  # shape (1, 2, 2)

# Calculate the difference between coordinates
difference = coords1_expanded - coords2_expanded  # shape (3, 2, 2)

# Square the differences
squared_difference = difference**2  # shape (3, 2, 2)

# Sum the squared differences along the last axis (the coordinate axis)
sum_of_squared_difference = np.sum(squared_difference, axis=-1)  # shape (3, 2)

# Take the square root to get the Euclidean distance
pairwise_distances = np.sqrt(sum_of_squared_difference)

print("Coordinates 1:")
print(coords1)
print("\nCoordinates 2:")
print(coords2)
print("\nPairwise Euclidean distances:")
print(pairwise_distances)
```

```python
#  48.How do you create a 2D Gaussian-like array?

# Define the size of the array
size = 5

# Create a meshgrid of coordinates
x, y = np.meshgrid(np.linspace(-1, 1, size), np.linspace(-1, 1, size))

# Calculate the 2D Gaussian function
# The formula for a 2D Gaussian centered at (0, 0) with standard deviation sigma is:
# exp(-(x^2 + y^2) / (2 * sigma^2))
sigma = 0.5
gaussian_array = np.exp(-(x**2 + y**2) / (2 * sigma**2))

print(gaussian_array)
```

```python
#  49.How do you randomly place a specific number of elements in a 2D array?

rows = 5
cols = 5

# Define the number of elements to randomly place
num_elements_to_place = 10

# Create a 2D array filled with zeros
arr = np.zeros((rows, cols), dtype=int)

# Get all possible indices in the 2D array
all_indices = np.arange(rows * cols)

# Randomly choose 'num_elements_to_place' indices without replacement
random_indices = np.random.choice(all_indices, size=num_elements_to_place, replace=False)

# Convert the flat indices to 2D indices
row_indices, col_indices = np.unravel_index(random_indices, (rows, cols))

# Place a specific value (e.g., 1) at the chosen random indices
arr[row_indices, col_indices] = 1

print(arr)
```

```python
#  69.How do you compute the distance from a point to multiple lines?


# Lines defined by two endpoints (N lines)
A = np.array([[0, 0], [1, 1], [2, 0]])   # starting points of lines
B = np.array([[1, 0], [2, 1], [3, 1]])   # ending points of lines

# The point P
P = np.array([1, 2])

# Vector along each line
BA = B - A

# Vector from A to the point P
PA = A - P

# Cross product magnitude (for 2D use determinant)
cross = np.abs(BA[:, 0] * PA[:, 1] - BA[:, 1] * PA[:, 0])

# Distance formula: |cross| / |BA|
distances = cross / np.linalg.norm(BA, axis=1)

print(distances)
```

```python
#  70.How do you compute the distances from multiple points to multiple lines?

# Lines: N lines
A = np.array([[0, 0], [1, 1], [2, 0]])   # start points
B = np.array([[1, 0], [2, 1], [3, 1]])   # end points

# Points: M points
P = np.array([[1, 2], [0, 1], [3, 3]])

# Step 1: Line vectors (shape: N x 2)
BA = B - A  # shape (N, 2)

# Step 2: Reshape points and lines for broadcasting
# P: M x 1 x 2, BA: 1 x N x 2, A: 1 x N x 2
PA = A[np.newaxis, :, :] - P[:, np.newaxis, :]  # shape: M x N x 2
BA = BA[np.newaxis, :, :]                        # shape: 1 x N x 2

# Step 3: Compute 2D cross product magnitude
cross = np.abs(BA[:, :, 0] * PA[:, :, 1] - BA[:, :, 1] * PA[:, :, 0])

# Step 4: Normalize by line length
line_lengths = np.linalg.norm(BA, axis=2)  # shape: 1 x N
distances = cross / line_lengths           # shape: M x N

print(distances)
```

```python
 # 71.How do you extract a fixed-shaped subarray centered on a given point?
arr = np.arange(100).reshape(10, 10)

# Center point
center_row, center_col = 5, 5

# Patch size
patch_height, patch_width = 3, 3

# Half sizes
h_half = patch_height // 2
w_half = patch_width // 2

# Compute start and end indices
row_start = max(center_row - h_half, 0)
row_end   = min(center_row + h_half + 1, arr.shape[0])
col_start = max(center_col - w_half, 0)
col_end   = min(center_col + w_half + 1, arr.shape[1])

# Extract patch
patch = arr[row_start:row_end, col_start:col_end]

print("Extracted Patch:\n", patch)
```

```python
#  79.How do you simulate the Game of Life using NumPy?

# Grid size
rows, cols = 5, 5
np.random.seed(0)  # for reproducibility

# Initialize random grid (0 = dead, 1 = alive)
grid = np.random.randint(2, size=(rows, cols))
print("Initial Grid:\n", grid)

# Number of generations to simulate
generations = 5

for g in range(generations):
    # Count neighbors using np.roll for 8 directions
    neighbors = (
        np.roll(grid, 1, 0) + np.roll(grid, -1, 0) +  # up, down
        np.roll(grid, 1, 1) + np.roll(grid, -1, 1) +  # left, right
        np.roll(np.roll(grid, 1, 0), 1, 1) +          # up-left
        np.roll(np.roll(grid, 1, 0), -1, 1) +         # up-right
        np.roll(np.roll(grid, -1, 0), 1, 1) +         # down-left
        np.roll(np.roll(grid, -1, 0), -1, 1)          # down-right
    )

    # Apply Game of Life rules
    new_grid = ((grid == 1) & ((neighbors == 2) | (neighbors == 3))) | \
               ((grid == 0) & (neighbors == 3))

    grid = new_grid.astype(int)  # convert boolean to int

    print(f"\nGeneration {g+1}:\n", grid)
```

```python
#  89.How do you sample a 2D path at regular intervals?
path = np.array([[0, 0],
                 [1, 2],
                 [3, 3],
                 [6, 1]])
# Differences between consecutive points
diffs = np.diff(path, axis=0)  # shape: (num_points-1, 2)

# Euclidean distance between consecutive points
distances = np.sqrt((diffs**2).sum(axis=1))

# Cumulative distance along the path
cumdist = np.insert(np.cumsum(distances), 0, 0)
num_samples = 10
regular_distances = np.linspace(0, cumdist[-1], num_samples)
x_regular = np.interp(regular_distances, cumdist, path[:, 0])
y_regular = np.interp(regular_distances, cumdist, path[:, 1])

sampled_path = np.stack([x_regular, y_regular], axis=1)
print(sampled_path)
```

```python
#  100.How do you detect outliers in a dataset using IQR?

data = np.array([10, 12, 14, 15, 18, 20, 22, 25, 90])
Q1 = np.percentile(data, 25)
Q3 = np.percentile(data, 75)
IQR = Q3 - Q1
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR
outliers = data[(data < lower_bound) | (data > upper_bound)]

print("Q1:", Q1)
print("Q3:", Q3)
print("IQR:", IQR)
print("Lower bound:", lower_bound)
print("Upper bound:", upper_bound)
print("Outliers:", outliers)
```
