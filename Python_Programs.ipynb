```python
# 2. Check if a year is a leap year. A leap year is a year that is divisible by 4, but if it is a
# century year (divisible by 100), it must also be divisible by 400 to be a leap year
year=int(input("Enter a year:"))
if year%4==0 and year%100==0:
  if year%400==0:
    print(year, "is a leap year")
  else:
    print(year, "is not a leap year")
elif year%4==0:
  print(year, "is a leap year")
else:
  print(year, "is not a leap year")
```

Mathematical Programs

```python
# 1.Check if a number is prime or not
number=int(input("Enter a number:"))
for i in range(2,number):
  if number%i==0:
    print(number,"is not a prime number")
    break
else:
  print(number,"is a prime number")
```

```python
# 2.  Find the sum of digits of a number (eg: input: 1234 and output: 10)
number=int(input("enter a number:"))
sum=0
while number>0:
  sum+=number%10
  number=number//10
print(sum)
```

```python
# 3. Find the LCM of two numbers
a=int(input("Enter first number:"))
b=int(input("Enter second number:"))
if(a>b):
  greter=a
else:
  greter=b
while True:
  if (greter%a==0 and greter%b==0):
    lcm=greter
    break
  greter+=1
print(lcm)
```

```python
#4  Find the GCD of two numbers
a=int(input("Enter first number:"))
b=int(input("Enter second number:"))
while b!=0:
  remainder=a%b
  a=b
  b=remainder
print("The GCD is ",a)
```

```python
# 5  Print Pascal's Triangle
def pascal_triangle(n):
    for i in range(n):
        # Print leading spaces for alignment
        print(" " * (n - i), end="")
        num = 1
        for j in range(i + 1):
            print(num, end=" ")
            num = num * (i - j) // (j + 1)
        print()  # Move to next line

rows = int(input("Enter number of rows: "))
pascal_triangle(rows)
```
